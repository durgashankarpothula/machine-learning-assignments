### What is the distribution  of preferred foot?

```python
distribution_of_preffered_foot=df["Preferred Foot"].value_counts()
distribution_of_preffered_foot
```

```python
distribution_of_preffered_foot.plot(kind="bar")
plt.xlabel("Distribution  of preferred foot")
plt.ylabel("Frequency")
plt.show()
```

### How can we visualize the age distribution of players?

```python
#example 1
plt.figure(figsize=(8, 11))
plt.hist(df["Age"])
plt.title("Age Distribution of Players")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#example 2
plt.figure(figsize=(11, 9))
sns.countplot(x=df["Age"],data=df)
plt.title("Age Distribution of Players")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#In the above graph we can see that most of the players are of age 21 & 26 
```

### What is the international reputation of left and right foot?

```python
df.groupby('Preferred Foot')['International Reputation'].value_counts()
```

```python
sns.countplot(x=df['International Reputation'], hue=df['Preferred Foot'],  data=df)
plt.title('International Reputation of Left and Right Footed Players ', fontsize = 12)
plt.show()
```

### Top 10 countries with most number of players

```python
most_number_of_players=df["Nationality"].value_counts().head(10) # assigning top 10 values by using head() function
most_number_of_players
```

```python
fig, ax = plt.subplots(figsize=(12,8))# defininig figure size
x = most_number_of_players.values
y = most_number_of_players.index
ax.barh(y, x, align='center', color='grey')
ax.set_title('Top 10 Countries with most number of players')
ax.set_xlabel('Number of Players')
ax.set_ylabel('Name of Countries')
plt.show()
```
